## Draw a line

x default to 0,1,2,3:

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot([1,2,3,4])
plt.ylabel('some numbers')
plt.show()
```

## Change Styles

```python
t = np.arange(.0, 5., 0.5)
line1, line2, line3 = plt.plot(t, t, 'ro', t, t**2, 'bs', t, t**3, 'g-')
line3.set_antialiased(False)
plt.show()
```

```python
t = np.arange(.0, 5., 0.5)
lines = plt.plot(t, t, 'ro', t, t**2.6, 'bs', t, t**3, 'g-')
plt.setp(lines, color='b', linewidth=3.0)
plt.setp(lines, 'color', 'r', 'linewidth', 2.0)
plt.show()
plt.setp(lines)
```

## 2+ subplots

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

## Histogram

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, normed=1, facecolor='g', alpha=0.75)

plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
note1 = plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.setp(note1, fontsize=20, color='red')
plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```

## Annotate on graph

```python
ax = plt.subplot(111)

t = np.arange(0.0, 5.0, 0.01)
s = np.cos(2*np.pi*t)
line, = plt.plot(t, s, lw=2)
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05))
plt.ylim(-2,2)
plt.show()
```

## Axes scale

```python
# make up some data in the interval ]0, 1[
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(1)

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)
# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)
# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.05)
plt.title('symlog')
plt.grid(True)
# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)

plt.show()
```
