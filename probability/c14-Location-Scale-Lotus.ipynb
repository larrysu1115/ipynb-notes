### Recap

$
E(Z) = 0 \\
Var(Z) = E(Z^2) = 1
$

$ E(Z^3) = 0 $, (called 3<sup>rd</sup> moment) because using LOTUS, this is an odd function:

$$
E(Z^3) = \int_{-\infty}^{\infty} z^3 \frac{1}{\sqrt{2 \pi}} e^{-z^2 / 2} dz
$$

$ -Z \sim N(0,1) $, by symmetry

# Scale, Location, $ X \sim N ( \mu, \sigma^2 ) $

Let $ X = \mu + \sigma Z $  
$ \mu \in \mathcal{R} $, called mean or location;  
$ \sigma \gt 0 $, called SD, scale

Then we say $ X \sim N ( \mu, \sigma^2 ) $

$ E(X) = E(\mu) + \underbrace{E(\sigma Z)}_{zero} = \mu $

$ Var(X) = Var( \mu + \sigma Z ) = Var( \sigma Z ) = \sigma^2 $

### Standardization:

$ Z = \frac{X - \mu}{\sigma} $

Note : 標準化後就沒有單位，例如原本 Normal 分佈是以 小時 或 秒 為單位，但經過 除以 sigma 後，單位就沒有了。

### More on Variance

由 variance 的定義可以知道，Var(X + c) = Var(X)

$
Var(X) = E \big( (X - EX)^2 \big) = EX^2 - (EX)^2 \\
Var(X + c) = Var(X) \\
Var(cX) = c^2 Var(X)
$

$ Var(X + Y) \ne Var(X) + Var(Y) $ in general. equal if X,Y are indep.  

for example $ Var(X+X) = Var(2X) = 4 Var(X) $

### Find PDF of $ N(\mu, \sigma^2) $

如果已經知道 PDF of Z, 如何知道 PDF of $ N(\mu, \sigma^2) $ ?

Let $ X \sim N(\mu, \sigma^2) $

first, find CDF: (using standardization)
$ 
P(X \le x) = P \big( \frac{X - \mu}{\sigma} \le \frac{x - \mu}{\sigma} \big) \\
= \Phi \big( \frac{x - \mu}{\sigma} \big)
$

PDF is derivatif of CDF, using chain rule:

$$
PDF : \frac{1}{\mu} \frac{1}{\sqrt{2 \pi}} e^{-(\frac{x-\mu}{\sigma})^2 / 2}
$$

Also:

$
-X = -\mu + \sigma( \underbrace{ -Z }_{\text{standard normal: -Z=Z}} ) \sim N(-\mu, \sigma^2)
$

Later will prove this:

if $ X_j \sim N(\mu_j, \sigma_j^2) $ are indep., then:

$$
X_1 + X_2 \sim N(\mu_1 + \mu_2, \sigma_1^2 + \sigma_2^2 ) \\
X_1 - X_2 \sim N(\mu_1 - \mu_2, \sigma_1^2 + \sigma_2^2 )
$$

注意相減時候，location 相減，但 scale 仍然是相加。$ X_1 - X_2 = X_1 + \underbrace{(-X_2)}_{N(-\mu, \sigma^2)} $

## 68-95-99.7 % RULE

因為不容易計算 $ \Phi $ (通常是查表),
這三個數指出了距離 mean 在 1,2,3 個 標準差SD之內 的機會

$$
P \big( \left| X - \mu \right| \le \sigma \big) \approx 0.68 \\
P \big( \left| X - \mu \right| \le 2 \sigma \big) \approx 0.95 \\
P \big( \left| X - \mu \right| \le 3 \sigma \big) \approx 0.997
$$

# LOTUS : Law of the Unconscious Statistician

在計算 VAR 時，需要用到 $ E(X^2) $, 可以利用 LOTUS 帶入。

$$
\text{ let } Y = X^2, E(X^2) = E(Y)
$$

need PDF of Y to calculate EY  
直接把 EX 的定義拿來平方:

$$
E(X) = \int_{-\infty}^{\infty} \ x \ f_X(x) \ dx 
\underbrace{\to}_{LOTUS} 
E(X^2) = \int_{-\infty}^{\infty} \ x^2 \ f_X(x) \ dx 
$$

In Continuous Case:

$$
E\Big( g(x) \Big) = \int_{-\infty}^{\infty} g(x) \ f_X(x) \ dx
$$

In Discrete Case:

$$
E\Big( g(x) \Big) = \sum_x g(x) \ P\big( X=x \big)
$$

### Explain

```
Prob: P0, P1, P2, P3,...
   X:  0,  1,  2,  3,...
 X^2:  0,  1,  4,  9,...
```

若 X 是 r.v., 可為 0,1,2,3,..., 為每個數字的對應機率是 $ P_0, P_1, P_2, \cdots $,  
可知道 $ X^2 $ 為 0,1,4,9,..., 為每個數字的對應機率也是 $ P_0, P_1, P_2, \cdots $,  
於是有 LOTUS 對 Expect Value 的替換發生。
