# Beta Distribution

$$
Beta \big(a, b\big), \ \ a \gt 0, \ \ b \gt 0
$$

### PDF:

$$
f \big( x \big) = c \ x^{a-1} (1-x)^{b-1}, \ \ \text{ for } 0 \lt x \lt 1, \ \ 0 \text{ otherwise } \\
c = \frac{(a+b-1)!}{(a-1)!\ (b-1)!}, \ \  \text{ prove later.}
$$

### Fexiable

flexiable family of continuous distribution.

if a=b=1, this is uniform.

if a=b= 1/2, this is U shape.

if a=b= 2, this is Cap shape.

often used as PRIOR for parameters between (0,1)

"Conjugate prior to Binomial"

connections to other distributions.

### Conjugate Prior for Binomial

$ X \ \Big| \ p \sim Bin(n,p) $

X: data to observe  
p: given probability p

$ p \sim Beta\big(a, b\big) $ [prior]

Find the posterior dist: $ p \ \Big| \ X $

$
f \big( p \ \big| \ X = k \big) 
= \frac{ P \big( X = k \ \big| \ p \big) \ f(p) }{P \big( X = k \big)}
= \binom{n}{k} p^k (1-p)^{n-k} \ \ \times 
\ \ c p^{a-1} (1-p)^{b-1} \ \Big/ P \big( X = k \big) \\
= \frac{c \binom{n}{k}}{P \big( X = k \big)} p^k (1-p)^{n-k} \ \ \times 
\ \  p^{a-1} (1-p)^{b-1} \\
\propto p^{a+k-1} (1-p)^{n+b-k-1} \\
\to p \ \big| \ X \sim Beta\big( a+X, \ n+b-X \big)
$

Conjugate Property : start with Beta distribution as PRIOR, calculate the POSTERIOR, POSTERIOR is still Beta.

If think  
X : # of successes
X : # of failures  
假設有事先已經有的資料 Beta(a,b), a個成功, b個失敗； 。
再加上 X 個成功, (n-X) 個失敗。就是:

$$
Beta \big( a + X \ , b + n - X \big)
$$


## Bayes' Billiards argument

$$ \int_0^1 c \ x^{a-1} (1 - x)^{b-1} dx = 1 $$

求解 c = ?

### Part 1 : 替換成類似 Binomial 的式子

$
\text{Let } \\
k=a-1 \\
b-1=n-k \\
\to \ n=a+b-2 
$

將原式轉換成非常類似 Binomial PDF : $ \binom{n}{k} p^k q^{n-k} $

$$
\int_0^1 c \ x^{a-1} (1 - x)^{b-1} dx \\
\to \int_0^1 c \ x^{k} (1 - x)^{n-k} dx \\
\to \frac{c}{\binom{n}{k}} \int_0^1 \ \binom{n}{k} \ \ x^{k} (1 - x)^{n-k} dx \\
$$

### Part 2 : Story of billiards.

case 1. 如有 n+1 顆白色撞球，將其中一顆塗成 粉紅色。然後將所有球排在 0到1 的直線上。r.v. X=粉紅球左邊的白球數

$
p \sim Unif(0,1) \ \leftarrow \text{ pink ball can be on any position between 0,1} \\
f(p) = 1
$

By law of total probability. 

$ P(B) = \sum_{i=1}^n \big( B \cap A_i \big) \ = \ \sum_{i=1}^n P(B|A_i)\ P(A_i) $

$$
P \Big(X=k \Big) = \int_0^1 P \Big(X=k \ \big| \ p \Big) \ f\big(p\big) dp
$$

若将左侧球视作 成功，左侧球加右侧球 平均分布在 (0,1) 区间上，中间被粉红球落在 p 上隔开，p 已经知道是介于 0~1 之间的小数  
所以是 左侧球成功 的几率 为 p, 是 右侧球失败 的几率 为 (1-p)，
左侧球是 k 颗，一共 左加右 侧球有 n 颗球。这就是一个 Bin(n,p) 的几率分布。

$$
P \Big( X = k \ \big| \ p \Big) = \binom{n}{k} \ p^{k} \ (1-p)^{n-k} \\
\to P \Big(X=k \Big) = \int_0^1 P \Big(X=k \ \big| \ p \Big) \ \underbrace{f\big(p\big)}_{\text{this is constanct:1}} dp \\
\to P \Big(X=k \Big) = \int_0^1 \binom{n}{k} \ p^{k} \ (1-p)^{n-k} dp
$$


case 2. 先將所有球排上 (0,1) 直線，再塗一顆球成 pink；這等同上例先塗再排。粉紅球左側有幾顆球 X=k 的機率，就是 1 / n+1；如 左側有1顆球，左側有2顆球... 機率都相同，只是看粉紅球落在哪裏。

$$
P \Big(X=k \Big) = \frac{1}{n+1}
$$

### Part 3 : 合併上述

合併上述:

$$
P \Big(X=k \Big) = \frac{1}{n+1} = \int_0^1 \binom{n}{k} \ p^{k} \ (1-p)^{n-k} dp
$$

在結合 Part 1 中想知道的:

$$
\int_0^1 c \ x^{a-1} (1 - x)^{b-1} dx \\
= \frac{c}{\binom{n}{k}} \int_0^1 \ \binom{n}{k} \ \ x^{k} (1 - x)^{n-k} dx \\
\to \frac{c}{\binom{n}{k}} \ \frac{1}{n+1} = 1 \\
\to c = \frac{(n+1)!}{k! \ (n-k)!} \\
\to c = \frac{(a+b-1)!}{(a-1)!\ (b-1)!}
$$
