# Covariance

def: 

$$
Cor\big(X,Y\big) = E \Big( (X-EX) (Y-EY) \Big)
$$

當兩數字相乘，兩數皆為正，乘數為正  
當兩數字相乘，兩數皆為負，乘數為正

如果 X=Y, Cor(X,Y) = Var(X) = $ E\Big( (X-EX)^2 \Big) $

Properties: 

$ Cor(X,X) = Var(X) $

$ Cor(X,Y) = Cor(Y, X) $

$ Cor(X,c) = 0 $, c is a constant.

*: $ Cor(cX,Y) = c \times Cor(X, Y) $

*: $ Cor(X,Y+Z) = Cor(X, Y) + Cor(X, Z) $

*: last 2 are also called "Bilinearity"

$ Cor\Big( \sum_{i=1}^m a_i X_i, \sum_{j=1}^n b_j Y_j \Big) = \sum_{i,j} a_i b_j \ Cor \big( X_i, Y_j \big) $

Property:

$$
Cor \big( X, Y \big) = E \big(XY\big) - E(X) E(Y)
$$

$$
Cor \big( X, Y \big) = E \Big( (X-EX) (Y-EY) \Big) \\
= E \Big( XY - Y \ EX - X \ EY + EX \ EY \Big) \\
= E(XY) - EY \cdot EX - EX \cdot EY + EX \cdot EY \\
= E \big(XY\big) - E(X) E(Y)
$$

E(Y EX) 中，EX為常數，所以為 $ EX \cdot E(Y) = EX \cdot EY $ 

### Cor, Sum

$ 
\begin{align}
Var \big(X_1 + X_2 \big) & = Cor \big( X_1 + X_2 \ , \ X_1 + X_2 \big) \\
& = Cor(X_1, X_1) + Cor(X_1, X_2) + Cor(X_2, X_1) + Cor(X_2, X_2) \\
& = Var(X_1) + 2 \times Cor(X_1, X_2) + Var(X_2)
\end{align}
$

If Cor is ZERO, then the "Variance of Sum" is "Sum of Variances".

If $ X_1, X_2 $ indep., then $ Cor(X_1, X_2) = 0, \ Var(X_1 + X_2) = Var(X_1) + Var(X_2) $

More generally:

$$
Var \Big(X_1 + \cdots + X_n \Big) = Var \big( X_1 \big) + \cdots + Var \big( X_n \big)
+ 2 \times \sum_{i<j} Cor\big( X_i, X_j \big)
$$

### Uncorrelated

Theorem: If X,Y are indep., then they're uncorrelated, i.e.: Cor(X,Y) = 0

CONVERSE is false.

example for converse: 

$
Z \sim N(0,1), \ \ X = Z, \ Y = Z^2
$

$
Cor(X,Y) = E(XY) - E(X) E(Y) = E(Z^3) - 0 = 0
$

X,Y are uncorrelated, but X,Y are very dependent!

# Correlation

Def:

$$
Corr \Big( X, Y \Big) = \frac{Cor\big( X, Y \big)}{SD\big( X\big) \ SD\big( Y\big)}
= Cor \Big( \frac{X - EX}{SD(X)}, \frac{Y - EY}{SD(Y)} \Big)
$$

也可以想作 : 將 X, Y 用 標準差 進行 normalization，成為 Standardized 後，再做 Corvariance.

Theorem:

inequality of Cauchy-Schwarz

$$
-1 \le Corr\big(X,Y\big) \le 1
$$

Proof:

WLOG (without lossing of generality) assume X,Y are standardized. (mean=0, var=1)

let $ Corr(X,Y) = \rho $

$
Var(X+Y) = Var(X) + Var(Y) + 2 Var(X,Y) = 1 + 1 + 2 \rho = 2 + 2 \rho \\
Var(X-Y) = Var \big( X + (-Y) \big) = Var(X) + Var(Y) - 2 Var(X,Y) = 2 - 2 \rho \\
Var(X) \ge 0 \\
2 + 2 \rho \ge 0 \to \rho \ge -1 \\
2 - 2 \rho \ge 0 \to \rho \le 1
$



### Cor in a Multinomial

$
\big( X_1, \cdots, X_k \big) \sim Mult \big(n, \vec{p} \big)
$

Find $ Cor \big( X_i, X_j \big) $ for all i, j

case 1: if $ i = j $, using Xi as binomial(n,p)

$ 
Cor(X_i, X_i) = Var(X_i) = n \ p_i \ ( 1 - p_i )
$

case 2: if $ i \ne j $

$
Cor \big( X_1, X_2 \big) = c \\
Var(X_1 + X_2) = n p_1 (1 - p_1) + n p_2 (2 - p_2) + 2c \\
Var(x_1 + X_2) = n \big( p_1 + p_2 \big) \big( 1 - (p_1 + p_2) \big) \\
Cor \big( X_1, X_2 \big) = - n p_1 p_2
$

General: $ \big( X_i, X_j \big) = - n \ p_i \  p_j $, for i not equal j.

### Variance of Binomial

$
X \sim Bin(n,p), \ \ as \ X = X_1 + X_2 + \cdots + X_n, \ \ X_j \ \ iid \ \  Bern(p)
$

Let $ I_A $ be indicator of r.v. of event A.

$ 
I_A^2 = I_A \\
I_A^3 = I_A
$

because $ I_A $ is 0 or 1.

$ I_A \ I_B = I_{A \cap B} = $ 0 or 1.

$ Var(X_j) = E X_j^2 - (EX_j)^2 = p - p^2 = p(1-p) $

$ Var(X) = npq $ 

因為 $ Cor(X_i, X_j) = 0 $ if i not equal j, they are iid, indep.

### Variance of HGeom(w, b, n)

$
X \sim HGeom \big(w, b, n \big) \\
p = \frac{w}{w+b} \\
w+b = N
$

$
X = X_1 + X_2 + \cdots + X_n, \ \ X_j = \begin{cases}
1 \text{ if jth ball is white } \\
0 \text{ otherwise }
\end{cases}
$

these are dependent indicator r.v.s

$
Var(X) = n Var(X_1) + 2 \binom{n}{2} Cor(X_1, X_2) \\
Cor(X_1, X_2) = E( X_1 X_2 ) - E(X_1) E(X_2) = \frac{w}{w+b} \frac{w - 1}{w+b -1} - \big( \frac{w}{w+b} \big)^2 \\
Var\Big( \sum_{j=1}^n X_j \Big) = Var(X_1) + \cdots + Var(X_n) + 
2 \sum_{i \lt j} Cor \big( X_i, X_j \big) \\
= \frac{N-n}{N-1} n \ p \ (1-p)
$

Finite Population Correction : $ \frac{N-n}{N-1} $

Examine with extreme cases:

n = 1 : just binomial(n,p)

N much larger than n : very close to binomial.
