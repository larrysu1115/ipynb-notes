### Theorem:

Let X be a continuous r.v. with PDF $ f_X, \ \ Y = g(X) $,  
where g is differentiable and strictly increasing.  
Then the PDF of Y is :

$$
f_Y \big(y\big) = f_X(x) \frac{dx}{dy} , 
\text{ where } y = g(x), \ x = g^{-1}(y) \\
\text{ and this is written in terms of y}
$$

Also:

$$
\frac{dx}{dy} = \big( \frac{dy}{dx} \big)^{-1}
$$

### Proof

find CDF, then take derivative to get PDF.

CDF of Y is: 

$
F_Y(y) = P\big( Y \le y \big) \\
= P \big( g(X) \le y \big) \\
= P \big( X \le g^{-1}(y) \big) = F_X \big( g^{-1}(y) \big)
= F_X \big(x\big)
$

using Chain Rule, PDF of Y is:

$$
f_Y(y) = \frac{d}{dy} F_Y(y)
= f_X(x) \frac{dx}{dy}
$$

f_Y(y) = 
\frac{1}{y} \

\text{ for } y \gt 0
$

$
\frac{dy}{dz} = e^z = y
$

### Example : transformation in $ \mathcal{R}^n $

$$
\vec{Y} = g \Big( \vec{X} \Big), \ \ g: \mathcal{R}^n \to \mathcal{R}^n \\
\vec{X} = \big( X_1, \cdots, X_n \big) \text{ is continuous }
$$

Find Joint PDF of $ \vec{Y} $ in terms of $ \vec{X} $

$$
f_{\vec{Y}} \big( \vec{y} \big) 
= f_{\vec{X}} \big( \vec{x} \big) \ 
\underbrace{\big| \frac{d\vec{x}}{d\vec{y}} \big|}_{\text{jacobian}}
$$

absolute value of determinant, Jacobian: 

$$
\frac{d\vec{x}}{d\vec{y}} = 
\begin{bmatrix}
\frac{dx_1}{dy_1} & \frac{dx_1}{dy_2} & \cdots & \frac{dx_1}{dy_n} \\
\vdots & & & \vdots \\
\frac{dx_n}{dy_1} & \frac{dx_n}{dy_2} & \cdots & \frac{dx_n}{dy_n}
\end{bmatrix} \\
\\
\big| \frac{d\vec{y}}{d\vec{x}} \big|^{-1} = \big| \frac{d\vec{x}}{d\vec{y}} \big|
$$

## Convolution (sums)

Let T = X + Y; X,Y are indep.

Find distribution of T.

#### in discrete case: 

$
P \Big( T = t \Big) = \sum_x P \Big( X = x \Big) \ P \Big( Y = t-x \Big)
$

#### in continuous case: 

$
f_T \Big( t \Big) = \int_{-\infty}^{\infty} f_X \Big( x \Big) \ f_Y \Big( t-x \Big) dx
$

since

$
F_T(t) = P\big( T \le t \big) \\
\int_{-\infty}^{\infty} P\big( X+Y \le t \ \big| \ X=x \big) \ f_X(x) \ dx \\ 
= \int_{-\infty}^{\infty} F_Y\big( t-x \big) \ f_X(x) \ dx
$

then take derivative of both sides.
