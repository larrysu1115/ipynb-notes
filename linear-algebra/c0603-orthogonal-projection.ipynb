## Unit 6.3 : Orthogonal Projection

Orthogonal Projection:  
The unique vector $ \vec{w} $ in subspace $ \mathcal{W} $ that is closest to vector $ \vec{u} $

We have $ \vec{u} = \vec{w} + \vec{z} $, where z is always orthogonal to all vectors in $ \mathcal{W} $.  
It is therefore worthy of studying **Orthogonal Complement** of $ \mathcal{W} $,  
the set of vectors that possess this property.

```python
%matplotlib inline
import matplotlib.pyplot as plt, math, numpy as np
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.set_xlim(0, 10)
ax.set_ylim(0, 10)
ax.set_zlim(0, 10)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
u = (np.array([2,7]), np.array([2,7]), np.array([1,10]))
w = (np.array([2,7]), np.array([2,7]), np.array([1,1]))
z = (np.array([2,2]), np.array([2,2]), np.array([1,10]))
W_x = np.arange(0.5,10,0.5)
W_y = np.arange(0.5,10,0.5)
W_x, W_y = np.meshgrid(W_x, W_y)

ax.plot(u[0], u[1], u[2], color='B', alpha=0.5)
ax.plot(w[0], w[1], w[2], color='R', alpha=0.5)
ax.plot(z[0], z[1], z[2], color='G', alpha=0.5)
ax.plot_surface(W_x, W_y, 1, color='Y', alpha=0.2, rstride=5, cstride=5)
ax.text(5,7,8.5, r'$ \vec{u} $', fontsize=20, alpha=0.7)
ax.text(6,8,1, r'$ \vec{w} $', fontsize=20, alpha=0.7)
ax.text(2,2,8, r'$ \vec{z} $', fontsize=20, alpha=0.7)
ax.text(9,9,1, r'$ \mathcal{W} $', fontsize=20, alpha=0.7);
```

### Definition : Orthogonal Complement

The **orthogonal complement** of a nonempty subset $ \mathcal{S} $ of $ \mathcal{R}^n $, denoted by $ \mathcal{S}^{\perp} $,  
is the set of all vectors in $ \mathcal{R}^n $ that are orthogonal to every vector in $ \mathcal{S} $. That is,  

$ \mathcal{S}^{\perp} = \{ v \in R^n : v \cdot u = 0, \ \exists u \in S \} $

### Example

$ \mathcal{S} = \mathcal{R}^n \to \ \ \mathcal{S}^{\perp} = \{ 0 \} $

$ \mathcal{S} = \{ 0 \} \to \ \ \mathcal{S}^{\perp} = \mathcal{R}^n $

若 S 是 x-y 平面，$ \mathcal{S}^{\perp} $ 為 z 軸

### Property

- $ \vec{0} \in \mathcal{S}^{\perp} $ for every nonempty subset S of R<sup>n</sup>
- if $ \vec{v}, \vec{w} \in \mathcal{S}^{\perp} $ for a nonempty subset S of R<sup>n</sup>, then $ \vec{v} + \vec{w} \in \mathcal{S}^{\perp} $ and $ c \vec{v} \in \mathcal{S}^{\perp} $ for any scalar c.
- The orthogonal complement of any nonempty subset of R<sup>n</sup> is a subspace of R<sup>n</sup>

### Proposition

- For any nonempty subset S of R<sup>n</sup>, $ (Span S)^{\perp} = S^{\perp} $
- Let W be a subspace of R<sup>n</sup>, and B be basis of W. Then $ B^{\perp} = W^{\perp} $


### Example

For $ \mathcal{W} = Span \Big \{ $
$ u_1 = \begin{bmatrix} 1\\1\\-1\\4 \end{bmatrix} $ , 
$ u_2 = \begin{bmatrix} 1\\-1\\1\\2 \end{bmatrix} $ 
$ \Big \} $

Find a basis $ \mathcal{B} \in W^{\perp} $

$ v \in W^{\perp} \iff u_1 \cdot v = u_2 \cdot v = 0 $  
$ v = [x_1 \ x_2 \ x_3 \ x_4]^T $ satisfies

$ x_1 + x_2 - x_3 + 4 x_4 = 0 $  
$ x_1 - x_2 + x_3 + 2 x_4 = 0 $


$ \begin{bmatrix} x_1\\x_2\\x_3\\x_4 \end{bmatrix} $ = 
$ \begin{bmatrix} -3 x_4 \\ x_3 - x_4 \\ x_3 \\ x_4 \end{bmatrix} $ = 
$ x_3 \begin{bmatrix} 0\\1\\1\\0 \end{bmatrix} $ + 
$ x_4 \begin{bmatrix} -3\\-1\\0\\1 \end{bmatrix} $ 

$ \mathcal{B} = \Big \{ $
$ \begin{bmatrix} 0\\1\\1\\0 \end{bmatrix} $ , 
$ \begin{bmatrix} -3\\-1\\0\\1 \end{bmatrix} $ 
$ \Big \} $ is a basis for $ W^{\perp} $

Let $ A = \begin{bmatrix} 1&1&-1&4 \\ 1&-1&1&2 \end{bmatrix} $,  
Then the above system of linear equation is equivalent to  
$ A \vec{x} = \vec{0} $  
W = Row A  
$ W^{\perp} = Null \ \ A $

### Property

For any real matrix A,  
$ (Row \ A)^{\perp} = Null \ A $, or  
$ (Col \ A)^{\perp} = Null \ A^T $

### Theorem 6.7 : Orthogonal Decomposition Theorem

Let $ W \subset R^n $ be a subspace of $ R^n $. Then, for any vector u in $ R^n $,  
There exist unique vectors $ w \in W $ and $ z \in W^{\perp} $ such that u = w + z.

In addition, if { $ v_1, v_2, \cdots, v_k $ } is an orthonormal basis for W, then

$ w = (u \cdot v_1) v_1 + (u \cdot v_2) v_2 + \cdots + (u \cdot v_k) v_k $

### Example

Let W be the solution space of $ x_1 - x_2 + 2 x_3 = 0 $,  
Then u = [ 1 2 4 ]<sup>T</sup> can be uniquely decomposed as u = w + z,  
where w $ \in W $ and $ z \in W^{\perp} $.  
How to find w and z ?

Find orthonormal basis B for W:

$ B = \Big \{ $
$ w_1 = \frac{1}{\sqrt{2}} \begin{bmatrix} 1\\1\\0 \end{bmatrix} $, 
$ w_2 = \frac{1}{\sqrt{3}} \begin{bmatrix} -1\\1\\1 \end{bmatrix} $
$ \Big \} $

$ w = ( u \cdot w_1 ) w_1 + ( u \cdot w_2 ) w_2 = [0, 4, 2] \in W $

$ z = u - w = [1, -1, 2] $


### Property

For any subspace W of R<sup>n</sup>,

$ dim W + dim W^{\perp} = n $

原因: 若 W = Row A<sup>k x n</sup>, rank A = k

$ W^{\perp} = Null \ A $  
$ dim W^{\perp} = Nullity \ of \ A = n - k $

### Definition : Orthogonal Projection

Let W be a subspace of R<sup>n</sup> and u in R<sup>n</sup>.  
The **Orthogonal Projection of u on W** is the unique vector w such that u - w $ \in W^{\perp} $.

The function $ U_W : R^n \to R^n $ such that U_W(u) is the orthogonal projection of u on W for every u $ \in R^n $  
is called the orthogonal projection operator on W.

### Example

延續上例，$ \vec{w} = U_W(\vec{u}) $

### Definition : Orthogonal Projection Matrix

The **standard matrix** of **orthogonal projection operator** $ U_W $ on a subspace W of $ R^n $  
is called the **Orthogonal Projection Matrix** for W and is denoted as $ P_W $

Question: 給定一個 subspace W, 如何找到 Orthogonal Projection Matrix ?

### Lemma

Let C be a matrix with linearly independent columns. Then $ C^T C $ is invertible

Proof: Suppose $ C^T C b = 0 $ for some b,  
$ \to \lVert C b \rVert^2 = (Cb) \cdot (Cb) = (Cb)^T Cb = b^T C^T C b = b^T = 0 $  
$ \to Cb=0 \to b = 0 $, since C has L.I. columns.  
Thus $ C^T C $ is invertible.

### Theorem 6.8

Let columns of matrix $ C_{n \times k} $ form a basis for a subspace W of R<sup>n</sup>.  
Then $ P_W = C \ \ (C^T C )^{-1} \ \ C^T $

Proof.

Let $ \vec{u} \in R^n $ and $ \vec{w} = U_W(\vec{u}) $  
since W = Col C,  
w = C v for some $ v \in R^k $ (v為用 basis C 進行線性組合成 w 的一組係數)  
and u - w $ \in W^{\perp} = (Col \ C)^{\perp} = (Row \ C^T)^{\perp} = Null \ C^T $.

$ \to 0 = C^T(u - w) = C^T u - C^T w = C^T u - C^T ( C v ) = 0 $  

$ \to C^T u = C^T ( C v ) $  
$ \to v = ( C^T C )^{-1} C^T u $  
$ \to \vec{w}_n = C \vec{v}_k = C ( C^T C )^{-1} C^T \vec{u}_n $

### Example

Let W be the solution space of $ x_1 - x_2 + 2 x_3 = 0 $,  
Then u = [ 1 2 4 ]<sup>T</sup> can be uniquely decomposed as u = w + z,  
where w $ \in W $ and $ z \in W^{\perp} $.  
How to find w and z ?

W has basis : $ B = \Big \{ $
$ \begin{bmatrix} 1\\1\\0  \end{bmatrix} $, 
$ \begin{bmatrix} -2\\0\\1 \end{bmatrix} $
$ \Big \} $

$ C = \begin{bmatrix} 1&-2 \\ 1&0 \\ 0&1  \end{bmatrix} $

使用公式計算 w, z 如下:

```python
import numpy as np
from numpy import linalg as la
np.set_printoptions(precision=3, suppress=True)
C = np.array([1,-2,1,0,0,1]).reshape(3,2)
u = np.array([1,3,4])
print('matrix C = \n%s\n' % C)
Pw = C.dot( la.inv(C.T.dot(C)) ).dot(C.T)
print('Orthogonal Projection Matrix Pw = \n%s\n' % Pw)
w = Pw.dot(u)
print('vector w = \n%s\n' % w)
z = u - w
print('vector z = \n%s\n' % z)
```
